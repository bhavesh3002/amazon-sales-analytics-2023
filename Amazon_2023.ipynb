```python
# Step 1: Keep original category name for reference
data['original_category'] = data['category_name']

# Step 2: Create sub_category based on original patterns
def extract_sub_category(name):
    name = str(name).lower()

    if 'shoe' in name or 'sandal' in name or 'ballerina' in name:
        return 'Footwear'
    elif 'bag' in name or 'backpack' in name or 'sack' in name or 'luggage' in name:
        return 'Bags'
    elif 'nightwear' in name or 'lingerie' in name or 'innerwear' in name or 'underwear' in name:
        return 'Innerwear'
    elif 'sport' in name or 'camping' in name or 'cricket' in name or 'badminton' in name or 'fitness' in name or 'cycling' in name:
        return 'Sports'
    elif 'toy' in name or 'game' in name or 'puzzle' in name or 'playstation' in name or 'xbox' in name:
        return 'Toys & Games'
    elif 'speaker' in name or 'appliance' in name or 'laptop' in name or 'tablet' in name or 'refrigerator' in name:
        return 'Electronics'
    elif 'home' in name or 'kitchen' in name or 'furniture' in name or 'bath' in name or 'lighting' in name:
        return 'Home Essentials'
    elif 'baby' in name or 'kid' in name or 'child' in name or 'diaper' in name:
        return 'Baby & Kids'
    elif 'beauty' in name or 'make-up' in name or 'hair' in name or 'skin care' in name:
        return 'Cosmetics'
    elif 'car' in name or 'automotive' in name or 'tire' in name or 'engine' in name:
        return 'Automotive'
    elif 'hardware' in name or 'tools' in name or 'equipment' in name or 'scientific' in name:
        return 'Industrial Tools'
    elif "clothing" in name:
        return 'Clothing'
    elif 'jewelry' in name or 'accessories' in name or 'watch' in name:
        return 'Accessories'
    elif 'pet' in name or 'dog' in name or 'cat' in name:
        return 'Pet Supplies'
    elif 'art' in name or 'paint' in name or 'craft' in name:
        return 'Arts & Crafts'
    elif 'stationery' in name or 'school' in name or 'office' in name:
        return 'Stationery'
    elif 'party' in name or 'seasonal' in name or 'wrapping' in name:
        return 'Seasonal & Party'
    elif 'health' in name or 'medical' in name or 'dental' in name:
        return 'Healthcare'
    else:
        return name.title()  # Keep original if no match

# Apply function to create sub_category
data['sub_category'] = data['original_category'].apply(extract_sub_category)
```

*Let's handle outliers now*

```python
def handle_outlier(data,column):
    sns.boxplot(data[column])
    plt.title("outlier detection")
    plt.show()
    q1=data[column].quantile(0.25)
    q3=data[column].quantile(0.75)
    iqr=q3-q1
    upper_limit= q3+1.5*iqr
    lower_limit=q1-1.5*iqr

    data[column]=np.where(
        data[column]>upper_limit, upper_limit, np.where(data[column]<lower_limit,lower_limit,data[column])
    )

    sns.boxplot(data[column])
    plt.title("outliers removed")
    plt.show()
```

```python
# Calculate the discount amount and percentage
data['discount_amount'] = data['price'] - data['listPrice']
data['discount_percentage'] = (data['discount_amount'] / data['price']) * 100
```

```python
# 2b) Average discount across categories
data.loc[data['discount_percentage']==-np.inf]
```

```python
#note that since price is zero.
#discount_percentage is getting infinity
```

```python
# Handle infinity
data['discount_percentage'] = data['discount_percentage'].replace([np.inf, -np.inf], 0)
# discount_percentage below 0 are set to 0.
data['discount_percentage'] = data['discount_percentage'].clip(lower=0)
```

```python
overall_average_discount_percentage=np.mean(data['discount_percentage'] )
print("Overall Average Discount Percentage is ",np.round(overall_average_discount_percentage,2))
```
